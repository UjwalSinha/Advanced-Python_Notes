```python
# Introduction to Matplotlib

# It is a visualization Library

# We can be able to showcase our data graphically by using the matplotlib
```

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt #using matplotlib as a library and from it considering pyplot module alisa name as 'plt'
```

```python
# req: Sales data
# Revenue vs sales
# x axis --- sales
# y axis --- Revenue
# title of graph  ---- Performance calculator of sales
```

```python
x = [1,2,3]
y = [1,4,9]

```

```python
plt.plot(x,y) # Shows graphical rep of data
plt.xlabel('sales') # labels x axis
plt.ylabel('Revenue') # labels y axis
plt.title('Performance calculator of sales') # labels chart title
plt.show() # prints the graph
```

```python
x = [1,2,3]
y = [1,4,9]
z = [10,5,0]
plt.plot(x,y) # This is x and y
plt.plot(x,z) # This is x and z
plt.xlabel('sales') # labels x axis
plt.ylabel('Revenue') # labels y axis
plt.title('Performance calculator of sales') # labels chart title
plt.legend(['this is x and y', 'this is x and z']) # Legend brings clarity to the graph
plt.show()
```
