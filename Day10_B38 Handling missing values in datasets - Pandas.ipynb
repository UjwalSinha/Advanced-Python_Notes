```python
# How to handle missing values in dataset
```

```python
import numpy as np
import pandas as pd
```

```python
# There are 2 approaches to handle missing data

# 1. get rid of rows and columns with nulls
# 2. Replace nulls with non-null values, this process is called imputation
```

```python
movies_df = pd.read_csv('IMDB-Movie-Data.csv', index_col = 'Title')
movies_df
```

```python
# Initial step we have to first understand how many columns in the dataset is having null values
```

```python
movies_df.isnull() #isnull for checking the null values in our dataset
```

```python
movies_df.isnull().sum() # finding the null values
```

```python
# Removing null values in rows
```

```python
movies_df.dropna() # This approach will be deleting atleast one null values in the rows
```

```python
x = movies_df.dropna() # after performing the operation, we need to assign that to a new variable to see the deleted null values.
```

```python
x.isnull().sum()
```

```python
# removing null values in columns
```

```python
movies_df.dropna(axis = 1) # column wise deleting the null values
```

```python
y = movies_df.dropna(axis = 1) # Need to assign a variable to see the deleted null data
```

```python
y.isnull().sum()
```

```python
# Filling the null values
```

```python
movies_df.isnull().sum() # Original data set with existing null values
```

```python
# I want to fill in the null values in revenue
```

```python
revenue = movies_df['Revenue (Millions)']
```

```python
revenue.head()
```

```python
# Req: fill up the blanks....
# Getting the mean and filling with that value...
```

```python
revenue_mean = revenue.mean()
revenue_mean
```

```python
# filling the null values
```

```python
revenue.fillna(revenue_mean, inplace=True)
```
